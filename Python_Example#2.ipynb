```python
def problem1(file):
    #importing regex,time and operating system libraries
    import re
    import os, os.path
    import time
    #input directory
    DIR = file
    #pattern to find new name of file
    pattern = 'Z\w+'
    #define variables
    count_changes = 0
    count_time = 0
    start = 0
    end = 0
    # find how many files with .pdbqt under input directory
    file_count = (len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name)) and re.search('.pdbqt',name)]))
    # find names of files with .pdbqt under input directory
    file_names = [name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name)) and re.search('.pdbqt',name)]
    for count in range(0,file_count):
        #open nth file
        f = open(os.path.join(DIR, file_names[count]),"r") 
        #take first line for new name
        line=f.readline().strip()
        #find new name with using regex and convert to string
        rename_code = ''.join(re.findall(pattern,line))
        #start time
        start = time.time()
        #rename old name to new name
        os.rename(os.path.join(DIR, file_names[count]),os.path.join(DIR, rename_code+'.pdbqt')) 
        #end time
        end = time.time()
        #count how many file changes
        count_changes +=1
        #count time
        count_time += (end-start)
        #close file
        f.close()
    print(f'the number of processed files {count_changes}')
    result_smalloutput = (count_time/count_changes)/3600
    result_bigoutput = result_smalloutput * 3000000
    print(f'estimated hours to process three million files:{result_bigoutput:.2f}')

def problem2(file):
    #importing regex,statistics libraries
    import re
    import statistics
    #define dictionary of input
    dict_blood = {}
    list_beforedict = []
    #define lists for min,max and average value of blood pressure
    press_avglist = []
    press_minlist = []
    press_maxlist = []
    #open file
    with open(file, 'r') as fh:
        for line in fh:
            line = line.rstrip("\n")
            #if line is null skip it
            if line == '':
                continue
            id_list = line.strip().split(',')
            list_beforedict.append(id_list)
        #drop column names in excel
        list_beforedict.pop(0)
        #for create dictionary from list
        for i in list_beforedict:
            #first column define for key of dict
            dict_blood[i[0]] = ''.join(i[1:])
        #using regex for take values from line, there are 3 groups
        pattern = '(\d+)\D+(\d+)-(\d+)'
        for item in dict_blood.keys():
            press_list = (re.match(pattern,dict_blood[item]))
            press_avglist.append(int(press_list.group(1)))
            press_minlist.append(int(press_list.group(2)))
            press_maxlist.append(int(press_list.group(3)))
        
        #display results
        print(f'the lowest and the highest blood pressures are {min((press_minlist))} {max((press_maxlist))} respectively\naverage of the mean value is {statistics.mean((press_avglist))}')
        #close file
        fh.close()
def problem3(file):
    #importing regex,statistics libraries
    import statistics
    import math
    #define a dictionary for keeping gdp per country
    dict_country_gdp = {} 
    #set dictionary key as Turkey
    dict_country_key = 'Turkey'
    #define variables
    gdp_list = []
    list_gdps = []
    results = []
    file_to_list = []
    decreased_year = []
    percentage_peryear = [0]
    #open file and read it
    with open(file,'r') as f:
        for line in f:
            #convert file lines to list
            file_to_list = line.strip().split(';')
            #setting key of dict from first value of line
            dict_country_gdp[file_to_list[0]] = file_to_list[1:]
        #find years    
        years_list = [k for k in dict_country_gdp['Country Name']]            
        #define list for keeping gdp values
        list_gdps = ([ k for k in dict_country_gdp[dict_country_key] ])
        #convert gdps string to float
        list_gdps = [float(k) for k in list_gdps]
    
        for gdp in range(0,len(list_gdps)-1):
            #compute percentages according to years 
            percentage_peryear.append(((list_gdps[gdp+1] - list_gdps[gdp] )  /   list_gdps[gdp]) * 100)
            #if there is a descrese,find year and apppend it to list
            if (list_gdps[gdp]-list_gdps[gdp+1]) > 0: #decrease
                decreased_year.append(int(years_list[list_gdps.index(list_gdps[gdp+1])]))
        #find max percentage in list and find year with index of value
        max_increase_year = years_list[percentage_peryear.index(max(percentage_peryear))]
                    
    print(f'the highest increase happened in the year {max_increase_year}')       
    print(f'years that GDP per capita decreased compared to previous years:\n{decreased_year}')

#problem4
class movie_analyzer:
    # instance attribute
    def __init__(self,file):
        #define variables
        lines = []
        genre_dict = {}
        #open file
        with open(file,'r') as f:
            for line in f:
                #split with special ch
                lines.append(line.strip().split('::'))
            for line in lines:
                #create dict
                genre_dict[line[1]] = line[2]
        #assign lines and genre_dict to self
        self.lines=lines
        self.genre_dict = genre_dict
    def freq_by_year(self,n):
        #import counter library
        from collections import Counter
        years = []
        for line in self.lines:
            #last value of dict is year
            years.append(line[-1])
            #compute n common year with Counter
        for year,value in Counter(years).most_common(n):
            print(f'{year}:{value}') 
    def top_genrelist(self,n):
        from collections import Counter
        genre_list = []
        for line in self.lines:
            #find genres and split them then extend to list
            genre_list.extend(line[-2].split('|'))
        for genre,value in Counter(genre_list).most_common(n):
            #compute n common genre with Counter
            print('{:<10s}:{:>5d}'.format(genre,value))

    def print_genre(self,key):
        #display genres of films
        print(self.genre_dict[key])
```
